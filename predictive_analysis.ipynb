# We'll use a basic dataset and apply linear regression, which is a fundamental predictive modeling technique.

# What is Predictive Analysis?

# Predictive analysis uses historical data to make predictions about future or unknown outcomes using statistical or machine learning models.

# Predicting House Prices

# We will:

# 1. Use a dataset with house size (in square feet) and price.

# 2. Train a linear regression model.

# 3. Predict price for a new house size.

```python
# Step 1: Import Required Libraries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Step 2: Sample Dataset

# House sizes (X) and prices (y)
X = np.array([500, 1000, 1500, 2000, 2500]) # Features
y = np.array([100000, 150000, 200000, 250000, 300000]) # Target
```

```python
# Step 3: 
# Create and Train the model

# Creating the Model
model = LinearRegression()

# Reshaping the Data
X = X.reshape(-1, 1) # Reshape X to be a 2D array with a single feature column

# Training the Model
model.fit(X, y)
```

```python
# Step 4: Make a prediction

# Predict price for a 1800 sq ft house
predicted_price = model.predict([[1800]])
print(f"Predicted price for 1800 sq ft: ${predicted_price[0]:,.2f}")
```

```python
# Step 5: Visualize the result

plt.scatter(X, y, color= 'blue', label= 'Actual Prices')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price ($)')
plt.title('Predictive Analysis: House Prices')
plt.legend()
plt.grid(True)
plt.show()
```

# Lessons:

# 1. X (features) are the inputs - here, house sizes.

# 2. y (target) is what we want to predict - house prices.

# 3. LinearRegression finds the best-fit line to predict new values.

# 4. We used model.predict() to forecast an outcome.
